```python
%%writefile tp1_parte3.py
from multiprocessing import Pipe
import os
import sys
DEFAULT_CHUNK_SIZE = 512
READ_BYTES = "rb"
WRITE_BYTES = "wb"


def validate_args(args):
    if len(args) < 2:
        print("Cantidad incorrecta de argumentos.")
        os._exit(1)

    path = args[1]
    if not os.path.isfile(path):
        print("La ruta al archivo no es valida.")
        os._exit(1)

    size = DEFAULT_CHUNK_SIZE
    if len(args) == 3:
        size = int(args[2])
        if size <= 0:
            print("TamaÃ±o del chunk invalido.")
            os._exit(1)        
    return (path, size)

def recieve_file_data(recieve, chunk_size, file_name):
    try:
        chunk = recieve.recv_bytes()
        with open(file_name, WRITE_BYTES) as f:
            while chunk:
                f.write(chunk)
                chunk = recieve.recv_bytes()
    except EOFError:
        recieve.close()

def send_file_data(send, path, chunk_size): 
    with open(path, READ_BYTES) as f:
        data = f.read(chunk_size)
        while data:
            send.send_bytes(data)
            data = f.read(chunk_size)
        f.close()
        send.close()

def main():
    path, chunk_size = validate_args(sys.argv)
    send, recieve = Pipe()
    pid = os.fork()
    if pid < 0:
        print("Error al crear el nuevo proceso")
        os._exit(1)
    if pid:
        # Proceso Hijo que recibe datos
        file_name = "copia_" + os.path.basename(path)
        send.close()
        recieve_file_data(recieve, chunk_size, file_name)
    else:
        # Proceso Padre que envia datos
        recieve.close()
        send_file_data(send, path, chunk_size)    

main()
```
